# Creation de la table Dim Produit 

```python
from pyspark.sql.functions import monotonically_increasing_id,expr

source_df = spark.sql("SELECT * FROM AmazingZoneLH.dbo_Product")
dimProduit_df = source_df.withColumn("Id_inc", expr("monotonically_increasing_id() + 1")).select("Id_inc", *source_df.columns)
display(dimProduit_df)
```

```python
delta_table_path = "Tables/dimProduit_nbk" #fill in your delta table path 
# data = spark.range(5,10) 
dimProduit_df.write.format("delta").mode("overwrite").save(delta_table_path)
```
